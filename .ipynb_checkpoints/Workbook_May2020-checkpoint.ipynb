### Data Loader Oxford Dataset 

```python
# Load Oxford Dataset and put it into standard format

with open('data/Oxford-IIIT/annotations/list.txt', 'r') as f:
    allLines = f.read().splitlines()

    
oxfordDataset = []
missingXML_oxford = []
    
for line in allLines: 
    if line[0]!='#':
        try:
            oxfordDataset_entry = line.split()
            tree = ET.parse('data/Oxford-IIIT/annotations/xmls/'+oxfordDataset_entry[0]+'.xml')
            root = tree.getroot()
            xmin = root.find('.//bndbox/xmin').text
            xmax = root.find('.//bndbox/xmax').text
            ymin = root.find('.//bndbox/ymin').text
            ymax = root.find('.//bndbox/ymax').text
            
            coordinates = [xmin,ymin,xmax,ymax]
     
            
            
            oxfordDataset.append(['data/Oxford-IIIT/images/'+oxfordDataset_entry[0]+'.jpg','-'.join(oxfordDataset_entry[0].lower().split("_")[:-1]),oxfordDataset_entry[1:3],coordinates])
          
        except FileNotFoundError: 
            missingXML_oxford.append(line.split()[0])
    
print(oxfordDataset)
```

###  Attempt to standardize categories to combine Data into one Dataset

```python
import json

def breeder(dataset):
    
    fullList = []
    
    for item in dataset:
        fullList.append(item[1])


    breedSet = set(fullList)
    
    return breedSet

unionDataset = breeder(stanfordDataset).union(breeder(oxfordDataset))

filename='cat_to_name.json'

if not os.path.exists(filename):
    start = 1
    outputDic = {}
    while len(unionDataset)!=0:
        outputDic[str(start)]=unionDataset.pop()
        start+=1
 #   print(outputDic)    
    with open('cat_to_name.json', 'w') as outfile:
        json.dump(outputDic,outfile)

with open(filename,'r') as infile:
    inputDic = json.load(infile)
        
print(max(map(int,inputDic.keys())))      

def standardizeIndex(dataset, dictionary):
    encoder={}
    for i in range(len(dictionary)):
        a = dictionary.pop(str(i+1))
        encoder[a]=str(i+1)
    
    for item in dataset:
        item[2][0]=str(encoder[item[1]])
        
    return 
    
standardizeIndex(oxfordDataset,inputDic.copy())
standardizeIndex(stanfordDataset,inputDic.copy())


```

```python
import random
%matplotlib inline

def setGenerator(dataset, perctrain=0.8):
        
    length = set(range(len(dataset)))
    index_train= random.sample(length,int(len(dataset)*perctrain))
    index_val= list(length - set(index_train))

    return ([dataset[i] for i in index_train],[dataset[j] for j in index_val])
    
def plotExamples(dataset):
    
    fig = plt.figure()
    
    ax = fig.add_subplot(1,1,1)
    
    toPlot = random.sample(dataset,1)
    xmin = float(toPlot[0][3][0])
    ymin = float(toPlot[0][3][1])
    width = float(toPlot[0][3][2])-xmin
    height = float(toPlot[0][3][3])-ymin
    ax.imshow(io.imread(toPlot[0][0]))
    ax.add_patch(ptch.Rectangle((xmin,ymin), width, height, angle=0.0, ec='b', lw=2.8,fill=False))
    ax.text(xmin+width, ymin, "goodboi" , color = 'b', fontdict=None)
    print(toPlot[0][3])
      
    return None
plotExamples(oxfordDataset)
plotExamples(stanfordDataset)


def plotFromDataLoader(boxOfPuppies,n=None):
    
    fig = plt.figure()
    
    ax = fig.add_subplot(1,1,1)
    
    toPlot = boxOfPuppies[random.randint(0,len(boxOfPuppies)-1)]
   
    if torch.is_tensor(toPlot['image']):
        
        image=toPlot['image'].numpy().transpose(1,2,0)
        bndbox=toPlot['bndbox'].numpy()
        
    else:
        image=toPlot['image']
        bndbox=toPlot['bndbox']
        


    xmin = float(bndbox[0])
    ymin = float(bndbox[1])
    width = float(bndbox[2])-xmin
    height = float(bndbox[3])-ymin
    ax.imshow(image)
    ax.add_patch(ptch.Rectangle((xmin,ymin), width, height, angle=0.0, ec='b', lw=2.8,fill=False))
    ax.text(xmin+width, ymin, "goodboi" , color = 'b', fontdict=None)
      
    return None

```

### Data Loader General 

```python
class boxOfPuppies(Dataset):
    
    def __init__(self,dataset,filename='cat_to_name.json',transforms=None):
        
        with open(filename,'r') as infile:
            self.inputDic = json.load(infile)
        
        self.dataset = dataset
        self.transforms = transforms
  #
        self.categs_breed = max(map(int,self.inputDic.keys()))
        self.categs_species = 3
        self.zeros_breed = np.zeros((self.categs_breed+1))
        self.zeros_species=np.zeros((self.categs_species+1))
        
        print(self.categs_species)
        print(self.zeros_species)
        return None
    
    def __len__(self):
        return len(self.dataset)
    
    def encodeOH(self,categs,bndbox):
        
        categs_ten = False
        bndbox_ten = False
        
        if torch.is_tensor(categs):
            categs = categs.numpy()
            categs_ten = True
        
        encodedOH_temp1 = self.zeros_breed.copy()
        encodedOH_temp1[int(categs[0])] = 1

        encodedOH_temp2 = self.zeros_species.copy()
        encodedOH_temp2[int(categs[1])] = 1
  
        
       # encodedOH = np.concatenate((encodedOH_temp1,encodedOH_temp2))
        encodedOH = encodedOH_temp1
     #   encodedOH = np.concatenate((encodedOH_temp1,bndbox))

        
        if categs_ten or bndbox_ten:
              encodedOH = torch.from_numpy(encodedOH).float()
        
        
        return encodedOH
    
    def extend(self,dataset):
        self.dataset = self.dataset + dataset
        return None
    
    
    def __getitem__(self,idx):

#        if torch.is_tensor(idx):
#            idx = idx.tolist()
        image = io.imread(self.dataset[idx][0])
        bndbox = np.array(self.dataset[idx][3])
        bndbox = bndbox.astype('float')
        categs = np.array([self.dataset[idx][2][0],self.dataset[idx][2][1]])
        categs = categs.astype('int')

        
        ##Encode for ML##
        
        
        
        sample = {'image':image, 'categs':categs, 'bndbox':bndbox}
    
        if self.transforms!=None:
            sample = self.transforms(sample)
            
        sample['encodedOH']=self.encodeOH(categs = sample['categs'],bndbox = sample['bndbox'])
        
        
        return sample


        return None
        
```

### Image Transformations

```python
class Rescale():
    
    def __init__(self, output_size):
        assert isinstance(output_size, (int, tuple))
        self.output_size = output_size
        
    def __call__(self, sample):
        
        image, categs, bndbox = sample['image'], sample['categs'], sample['bndbox']
     
        h,w = image.shape[:2]
        
        
        if isinstance(self.output_size,int):
            if h>w:
                new_h, new_w = self.output_size * h/w, self.output_size
            else:
                new_h, new_w = self.output_size, self.output_size* w / h
        else:
            new_h, new_w = self.output_size
        
        new_h, new_w = int(new_h), int(new_w)
     
        image_out = transform.resize(image, (new_h, new_w))
        
        bndbox_out = bndbox.copy()
        bndbox_out[0]=int(bndbox[0]*new_w/w)
        bndbox_out[1]=int(bndbox[1]*new_h/h)
        bndbox_out[2]=int(bndbox[2]*new_w/w)
        bndbox_out[3]=int(bndbox[3]*new_h/h)
  
       
   
      
        return {'image':image_out, 'categs':categs,'bndbox':bndbox_out}
    
class RandomCrop():
    
    def __init__(self, output_size):
        assert isinstance(output_size,(int,tuple))
        if isinstance(output_size, int):
            self.output_size = (output_size, output_size)
        else:
            assert len(output_size)==2
            self.output_size = output_size
            
    def __call__(self,sample):
        image, categs, bndbox = sample['image'], sample['categs'], sample['bndbox']
        
        h,w = image.shape[:2]
        
        
        new_h, new_w = self.output_size
        
     
        top = random.randint(0, h-new_h)
        left= random.randint(0, w-new_w)
        
        bndbox_out = bndbox.copy()
        bndbox_out[0]=bndbox[0]-left
        bndbox_out[1]=bndbox[1]-top
        bndbox_out[2]=bndbox[2]-left
        bndbox_out[3]=bndbox[3]-top
  

    
        image_out = image[top: top+new_h, left:left+new_w]
        

        
        return {'image':image_out,'categs':categs, 'bndbox':bndbox_out}
    
class ToTensor():
        
    def __call__(self,sample):
        image, categs, bndbox = sample['image'], sample['categs'], sample['bndbox']
            
        image = image.transpose((2,0,1))
        
                # swap color axis because
                # numpy image: H x W x C
                # torch image: C X H X W
                        
        return {'image':torch.from_numpy(image).float(),'categs':torch.from_numpy(categs).float(), 'bndbox':torch.from_numpy(bndbox).float()}


class Normalize():
    
    def __init__(self,mean=[0.485, 0.456, 0.406],std=[0.229, 0.224, 0.225]):
    
        self.mean = mean
        self.std = std
    
    def __call__(self,sample):
        
        image = sample['image']
        image_out = image.copy()
        
        for i in range(0,3):
            image_out[i] = image[i]/self.mean[i]*self.std[i]
            
            
        
        
        return {'image':image_out, 'categs':sample['categs'], 'bndbox':sample['bndbox']}

    
```

```python
def dynDevice():
    if torch.cuda.device_count() == 0:
        device = 'cpu'
    elif torch.cuda.device_count() > 0:
        device = 'cuda'
        nDevice = torch.cuda.device_count()  #setup for later implementation of multi gpu parallel processing
    return device

print(dynDevice())


torch.cuda.empty_cache()

```

```python
neuralModel = torchvision.models.resnext50_32x4d(pretrained = True)

output_vec_len = 145

print(neuralModel)

for param in neuralModel.parameters():
    param.requires_grad = False    

classifier = nn.Sequential(OrderedDict([
                          ('fullyconnectedlayer1', nn.Linear(2048, 1024)),    #unclear how to determine size of input tensor
                          ('relu1', nn.ReLU()),
                          ('drop1', nn.Dropout(0.22)),
                          ('fullyconnectedlayer2', nn.Linear(1024, 512)),
                          ('relu2', nn.ReLU()),
                          ('drop2', nn.Dropout(0.22)),
                 #         ('fullyconnectedlayer3', nn.Linear(1024, 512)),
                 #         ('relu3', nn.ReLU()),
                 #         ('drop3', nn.Dropout(0.15)),
                 #         ('fullyconnectedlayer4', nn.Linear(512, 256)),
                 #         ('relu4', nn.ReLU()),
                 #         ('drop4', nn.Dropout(0.15)),
                          ('fullyconnectedlayer5', nn.Linear(512, output_vec_len)),
                          ('output', nn.LogSoftmax(dim=1))
                          ]))

neuralModel.fc = classifier

print(neuralModel)
```

```python
criterion = nn.NLLLoss()
optimizer = optim.AdamW(neuralModel.fc.parameters(), lr=0.0001, betas=(0.9, 0.999), eps=1e-08, weight_decay=0.01, amsgrad=True)
#learnRate_optimizer = optim.lr_scheduler.StepLR(optimizer, step_size=4, gamma=0.1)
#optimizer = optim.SGD(neuralModel.classifier.parameters(), lr=0.01, momentum=0.9) #experimenting with this
epochs = 30
generations=[]
training_losses, valid_losses, valid_accuracies = [], [], []
torch.cuda.empty_cache()
neuralModel.to(dynDevice())


for i in range(epochs):
    #Train
    neuralModel.train()
    optimizer.zero_grad()
    training_loss = 0
    start_time = time.time()

    for sample in train_loader:
        inputs,labels = sample['image'],sample['encodedOH']
        labels = torch.max(labels, 1)[1]
        inputs, labels = inputs.to(dynDevice()), labels.to(dynDevice())
        outputs = neuralModel.forward(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        training_loss += loss.item()

    timer_train = time.time() - start_time

    #Validation 
    neuralModel.eval()
    valid_loss = 0
    accuracies = 0
    start_time = time.time()

    with torch.no_grad():
        for sample in valid_loader:
            inputs,labels = sample['image'],sample['encodedOH']
            labels = torch.max(labels, 1)[1]

            inputs, labels = inputs.to(dynDevice()), labels.to(dynDevice())
            outputs = neuralModel.forward(inputs)

            #Calculate Validation Loss
            loss = criterion(outputs, labels)
            
            valid_loss += loss.item()

            #Calculate Accuracy 
            top_perc, top_cat = outputs.topk(1, dim=1)
 #           print(top_perc.shape)
 #           print(top_cat.shape)
 #           print(labels.shape)
 #           print(top_perc)
 #           print(top_cat)
 #           print(labels)
 #           break
            accuracy_equals = top_cat == labels.reshape(*top_cat.shape)
            accuracy = torch.mean(accuracy_equals.type(torch.FloatTensor))
            accuracies += accuracy.item()

    timer_valid = time.time() - start_time
    training_losses.append(training_loss/len(train_loader))
    valid_losses.append(valid_loss/len(valid_loader))
    valid_accuracies.append(accuracies/len(valid_loader))

    print(f'The {i+1}. epoch just finished. \n \
    The training loss was:{training_loss/len(train_loader):.3f}; \
    The validation loss was: {valid_loss/len(valid_loader):.3f}; \n \
    Training took {timer_train:.3f}s;   Validation took {timer_valid:.3f}s; \n \
    The accuracy was {accuracies/len(valid_loader)*100:.3f} Percent')

else: 
    generations.append(epochs)
```

```python
# Test
test_loss = 0
neuralModel.to(dynDevice())
neuralModel.eval()
start_time = time.time()
test_accuracies = []
accuracies = 0

with torch.no_grad():
    for inputs, labels in valid_loader:
        inputs, labels = inputs.to(dynDevice()), labels.to(dynDevice())
        output_logs = neuralModel.forward(inputs)
        loss = criterion(output_logs,labels)
        test_loss+=loss.item()
        outputs = torch.exp(output_logs)
        top_perc, top_cat = outputs.topk(1, dim=1)
        accuracy_equals = top_cat == labels.reshape(*top_cat.shape)
        accuracy = torch.mean(accuracy_equals.type(torch.FloatTensor))
        accuracies+= accuracy.item()        

    test_accuracies.append((f'After {sum(generations)} epochs',accuracies/len(test_loader)))
    timer_test = time.time() - start_time
    
    print(f'The Test loss was:{test_loss/len(test_loader):.3f} after {sum(generations)} training epochs; \n \
    Running the model on the {dynDevice()} took {timer_test:.3f}s; \n \
    The accuracy was {accuracies/len(test_loader)*100:.3f} Percent')
```

```python
def writeItDown(model, optimizer, class_to_idx, criterion = None, scheduler = None, arch="vgg19", path = "checkpoint_auto.pth", 
                generations = [0], valid_losses=[0],training_losses=[0],valid_accuracies = [0],test_accuracies = [0]):
    
    model_state_dict = model.state_dict()

    running_variables = {"generations":generations, 
                            "training_losses":training_losses,
                             "valid_losses":valid_losses,
                             "valid_accuracies":valid_accuracies, 
                             "test_accuracies":test_accuracies}
      
    if criterion == None:
        criterion = nn.NLLLoss()
        
    optimizer_state_dict = optimizer.state_dict
    #!!!is there some redundance in this 'savefile'? do I need to both save the model and the model_state_dic? !!!
    #!!! could I just save the model and then, when loading, load the statedict by "loadedModel.loadStateDict(loadedModel.state_dict)"?
    savefile = {"model":model, 
                    "model_state_dict":model_state_dict, 
                    "class_to_idx":class_to_idx, 
                    "criterion":criterion, 
                    "arch":arch, 
                    "classifier":model.fc,
                    "optimizer":optimizer, 
                    "optimizer_state_dict":optimizer_state_dict, 
                    "scheduler":scheduler, 
                    "running_variables":running_variables}
    torch.save(savefile, path)
    return None 

writeItDown(neuralModel, 
                optimizer = optimizer, 
                class_to_idx=inputDic, 
                criterion = criterion,
                arch='resnext50_32x4d', 
                generations = generations,
                training_losses = training_losses,
                valid_losses=valid_losses,
                valid_accuracies = valid_accuracies)
```
